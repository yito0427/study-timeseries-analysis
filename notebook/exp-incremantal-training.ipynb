# 時系列データセットを作る

```python
import pandas as pd
import numpy as np
```

## 定数のデータ

https://qiita.com/wrblue_mica34/items/2cfe71fa83f1c2bc7895


```python
pd.date_range(start='2018-2-1', periods=312, freq='H') #開始と期間，単位を指定，1時間
```

```python
df = pd.DataFrame({'item_id' : 'const',
                   'timestamp' : pd.date_range(start='2018-2-1', periods=312, freq='H'), #開始と期間，単位を指定，1時間
                   'demand' : np.full(312, 100)
                  })
```

```python
df
```

```python
df.to_csv('../input_for_forecast_const/tts_const.csv', index=False, header=False)
```

# 追加用データセット作成
・データセットに追加して、増分学習をかけることができるか検証する。  
・horizonは48Hとする。  

```python
df_add = pd.DataFrame({'item_id' : 'const',
                   'timestamp' : pd.date_range(start='2018-2-14', periods=24, freq='H'), #開始と期間，単位を指定，1時間
                   'demand' : np.full(24, 1000)
                  })
```

```python
df_add
```

```python
df_add.to_csv('../input_for_forecast_const/tts_const_add24h.csv', index=False, header=False)
```
